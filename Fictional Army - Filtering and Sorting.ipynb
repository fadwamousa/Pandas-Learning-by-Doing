```python
import pandas as pd
```

```python
# Create an example dataframe about a fictional army
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
            'deaths': [523, 52, 25, 616, 43, 234, 523, 62, 62, 73, 37, 35],
            'battles': [5, 42, 2, 2, 4, 7, 8, 3, 4, 7, 8, 9],
            'size': [1045, 957, 1099, 1400, 1592, 1006, 987, 849, 973, 1005, 1099, 1523],
            'veterans': [1, 5, 62, 26, 73, 37, 949, 48, 48, 435, 63, 345],
            'readiness': [1, 2, 3, 3, 2, 1, 2, 3, 2, 1, 2, 3],
            'armored': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0, 1, 1],
            'deserters': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
            'origin': ['Arizona', 'California', 'Texas', 'Florida', 'Maine', 'Iowa', 'Alaska', 'Washington', 'Oregon', 'Wyoming', 'Louisana', 'Ge']}
```

```python
df = pd.DataFrame(raw_data)
```

```python
df
```

```python
df.set_index('origin' , inplace = True)
```

```python
df
```

```python
df['veterans'] #Print only one column
```

```python
df[['deaths','veterans']] #Print two or more columns
```

```python
df.columns
```

```python
df.loc[['Maine','Alaska'],['deaths','size','deserters']] # Use Loc to access the data by String Index
```

```python
df.iloc[3:6 , 3:7] #Use Iloc to access the data by Int Index (Rows,Columns)
```

```python
df.iloc[3: , :] # All rows after the third Rows and all columns
```

```python
df.iloc[:3 , :] # All rows before the third Rows and all columns
```

```python
df.iloc[: , 3:7] # All columns from 3 to 7 and all rows
```

```python
df[df['deaths'] > 50]
```

```python
df[(df['deaths'] > 500) | (df['deaths'] > 500)]
```
